```python
import pandas as pd
import re
import os

# Leer el archivo Excel
df = pd.read_excel("metadata.xlsx")
df

```

```python
# Remove duplicates based on the 'title' column
metadata_unique_df = df.drop_duplicates(subset=["reduced_title"])
```

```python
metadata_unique_df
```

```python
def extract_base_title(title):
    """
    Limpia el título para extraer el nombre base, eliminando información de episodios y temporadas,
    incluyendo patrones como T:1 Ep:01 y otros similares.
    """
    # Convertir el título a string si no lo es, manejar valores faltantes
    if pd.isnull(title):  # Verificar si es un valor nulo
        return ''
    title = str(title)  # Convertir a string si no lo es

    # Expresión regular para eliminar patrones de temporadas/episodios
    # Maneja patrones como T:1, T01, E:10, Ep:11, Temporada 1 Episodio 1, etc.
    clean_title = re.sub(
        r'(T:? ?\d{1,3}|E:? ?\d{1,3}|\d{1,2}/\d{1,2}|Ep:? ?\d{1,3}|S\d+E\d+|\d+x\d+|Temporada\s*\d+\s*Episodio\s*\d+)',
        '',
        title,
        flags=re.IGNORECASE
    )

    # Eliminar caracteres extra y espacios redundantes
    clean_title = re.sub(r'[^A-Za-z0-9\s]', '', clean_title).strip()
    return clean_title

```

```python
# Aplicar la función a la columna 'title' y crear una nueva columna
df['Base_Title'] = df['title'].apply(extract_base_title)

# Mostrar los primeros resultados
print(df[['title', 'Base_Title']].tail())
```
